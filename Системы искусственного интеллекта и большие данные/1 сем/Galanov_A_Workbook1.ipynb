```python
import math
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapezoid

a = []
for x in range(0, 11):
  a.append(abs(math.cos(x * math.e ** (math.cos(x) + math.log(x + 1, math.e)))))

x = list(range(0 ,11))
y = a
plt.grid()
plt.plot(x, y, c = 'r')
plt.fill_between(x, y, alpha = 0.25)

area = trapezoid(y)
print(area)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

apple =    [131, 121, 122, 131, 124, 136, 145, 151, 141, 149, 165, 177]
microsoft = [222, 232, 235, 252, 249, 270, 284, 301, 281, 331, 330, 336]
google =   [1751, 2095, 2062, 2358, 2410, 2529, 2704, 2904, 2673, 2960, 2921, 2893]

plt.figure(figsize=(12, 6))
plt.plot(months, apple, marker='o', label="Apple")
plt.plot(months, microsoft, marker='o', label="Microsoft")
plt.plot(months, google, marker='o', label="Google")

plt.title("  Apple, Microsoft  Google  2021 .")
plt.xlabel("")
plt.ylabel(" (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import math

x = float(input(" x: "))
y = float(input(" y: "))

print("\n :")
print("+   ")
print("-   ")
print("*   ")
print("/   ")
print("exp  e^(x+y)")
print("sin  sin(x+y)")
print("cos  cos(x+y)")
print("pow  x**y")

op = input("\n: ")

if op == "+":
    print(":", x + y)
elif op == "-":
    print(":", x - y)
elif op == "*":
    print(":", x * y)
elif op == "/":
    if y != 0:
        print(":", x / y)
    else:
        print(":   ")
elif op == "exp":
    print(":", math.exp(x + y))
elif op == "sin":
    print(":", math.sin(x + y))
elif op == "cos":
    print(":", math.cos(x + y))
elif op == "pow":
    print(":", x ** y)
else:
    print(" ")
```
